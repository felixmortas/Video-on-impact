```python
import pandas as pd
from scipy import stats
```

```python
df = pd.read_excel('Video-on_4G_consumption.xlsx')
```

```python
df.head()
```

```python
# Calculate data consumption per minute for each call
df['Data Consumption per Minute'] = df.groupby('Call ID')['4G Consumption (Go)'].transform('mean') / df.groupby('Call ID')['Duration (minutes)'].transform('mean')
```

```python
# Group the data by camera usage and calculate the average data consumption per minute
camera_on = df[df['Time With Camera (minutes)'] > 0]['Data Consumption per Minute'].mean()
camera_off = df[df['Time Without Camera (minutes)'] > 0]['Data Consumption per Minute'].mean()
```

```python
# Perform a t-test to determine if the difference is statistically significant
t_stat, p_value = stats.ttest_ind(
    df[df['Time With Camera (minutes)'] > 0]['Data Consumption per Minute'],
    df[df['Time Without Camera (minutes)'] > 0]['Data Consumption per Minute']
)

print(f'Average data consumption per minute with camera on: {camera_on} Go')
print(f'Average data consumption per minute with camera off: {camera_off} Go')
print(f'T-statistic: {t_stat}')
print(f'P-value: {p_value}')
```
